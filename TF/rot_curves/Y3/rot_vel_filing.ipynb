## Marking all 'bad' velocity maps as 1 to not use in curve fitting

```python
from astropy.table import Table
from tqdm import tqdm
```

```python
loa_rotvel = Table.read('/pscratch/sd/d/dbustos/rot_curves/loa_rot_velocity.fits')
loa_rotvel[:5]
```

```python
bad_maps = Table.read('rot_curve_filing.txt', format = 'ascii.commented_header')
```

```python
bad_fibers = [2846575873228802, 2851283740393475, 2847201055211522, 2847362141650947,
              2851490922233858, 2407228879405058, 2407238362726401, 2851271849541651,
              2851184255696899, 2390684958982144, 2398375903232002, 2411570042765314,
              2851374693875714, 2846856132427781, 2394348310233105, 2407542449766417,
              2847230633443335]
```

```python
loa_rotvel['bad_map']=0
```

```python
for i in tqdm(bad_maps['SGA_ID']):
    loa_rotvel['bad_map'][loa_rotvel['SGA_ID']==i]=1
```

```python
for i in tqdm(bad_fibers):
    loa_rotvel['bad_map'][loa_rotvel['TARGETID']==i]=1
```

```python
loa_rotvel.write('/pscratch/sd/d/dbustos/rot_curves/loa_rotvel.fits',format='fits',overwrite=True)
```
