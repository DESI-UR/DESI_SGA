# Add morphology to shredded galaxies files
using Julia's SGA2020 morphologies as of 07/16/2026

```python
from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from astropy.io import fits

from astropy.coordinates import SkyCoord
import astropy.units as u

# from scipy.spatial import KDTree 
from tqdm import tqdm

from astropy.visualization.wcsaxes import SphericalCircle


from get_cutouts import *

import os
```

## load data

```python
data_directory = '/pscratch/sd/n/nravi/fastspecfit_shredded/'
morphology_fn = '/global/cfs/cdirs/desi/science/td/pv/SGA2020/final_classifications_SGA2020.csv'
```

```python
morphology = Table.read(morphology_fn)
morphology[:5]
```

```python
morph_dict = {}
for i in range(len(morphology)):

    morph_dict[int(morphology['SGA_ID'][i])] = morphology['Classification'][i]
```

```python
labels = {'Elliptical' : -1,
          'Lenticular': 0,
          'Spiral': 1,
          'Irregular' : -99}
```

## run through shredded galaxy files and add morphology

key: -1 elliptical, 0 lenticular, 1 spiral, 2 irregular, -99 unclassified

```python
fs_file_list = os.listdir(data_directory)
```

```python
for fn in tqdm(fs_file_list):

    table_fn = data_directory + fn

    t = Table.read(table_fn)
    t['MORPH'] = np.ones(len(t), dtype=int)*-99

    for i in range(len(t)):
        SGA_ID = t['SGA_ID'][i]
        try:
            t['MORPH'][i] = labels[morph_dict[SGA_ID]]
        except:
            continue

    t.write(table_fn, overwrite=True)
```
