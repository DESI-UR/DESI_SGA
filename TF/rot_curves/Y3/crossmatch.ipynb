```python
from astropy.table import Table, join, hstack
from astropy.coordinates import SkyCoord, match_coordinates_sky
from astropy import units as u

import os

import numpy as np
```

## Catalog crossmatching with SGA galaxies
len(SGA) = 383620

```python
SGA = Table.read('/global/cfs/projectdirs/cosmo/data/sga/2020/SGA-2020.fits', 'ELLIPSE')
SGA[:5]
```

```python
SGA_coords = SkyCoord(ra = SGA['RA'] * u.deg, dec = SGA['DEC'] * u.deg)
```

### CIGALE crossmatch

```python
CIGALE = Table.read('/global/cfs/cdirs/desicollab/users/zouhu/vac/dr2/dr2_galaxy_sedfitting_v1.0.fits')
CIGALE[:5]
```

```python
cig_coords = SkyCoord(ra = CIGALE['TARGET_RA'], dec = CIGALE['TARGET_DEC'])
```

```python
#index, 2d separation, 3d separation
idx, d2d, d3d = match_coordinates_sky(SGA_coords, cig_coords)
```

```python
#get only the targets within 1 arcsec
max_sep = 1 * u.arcsec
sep_constraint = d2d < max_sep

sga_matches = SGA[sep_constraint]
cat = CIGALE[idx[sep_constraint]]
```

203976 matches

```python
# sga_new1 = hstack([sga_matches['SGA_ID'], cat['TARGET_RA'], cat['TARGET_DEC'], cat['MASS_CG_5'], cat['MASSERR_CG_5']])
# make a new table with SGA_ID, mass, and mass error
cig_match = hstack([sga_matches['SGA_ID'], cat['MASS_CG_5'], cat['MASSERR_CG_5']])
cig_match[:5]
```

```python
# match SGA_ID in match_table to SGA catalog, and add mass + mass error to SGA table
sga_cig = join(SGA, cig_match, keys = 'SGA_ID', join_type = 'left')
sga_cig[:5]
```

```python
sga_cig.write('/pscratch/sd/d/dbustos/rot_curves/sga_xmatch.fits',format='fits', overwrite=True)
```

## H1 Survey Crossmatch

```python
# read in SGA alfalfa index crossmatch
SGA_a = Table.read('/pscratch/sd/d/dbustos/rot_curves/SGA-2020_ALFALFA.fits')
SGA_ALFALFA = SGA_a['SGA_ID', 'ALFALFA_index']
SGA_ALFALFA['ALFALFA_index'].name = 'ALFALFA_AGCNr'
SGA_ALFALFA[:5]
```

```python
# join index into SGA table
SGA = join(SGA, SGA_ALFALFA)
SGA[:5]
```

```python
ALFALFA = Table.read('/pscratch/sd/d/dbustos/rot_curves/a100.code12.table2.190808.csv',
                     format='ascii.csv',
                    include_names = ['AGCNr', 'logMH', 'siglogMH'])

ALFALFA[:5]
```

```python
ALFALFA_dict = {}

for i in range(len(ALFALFA)):
    
    ALFALFA_dict[ALFALFA['AGCNr'][i]] = i
```

```python
SGA['logMH_alf'] = np.nan
SGA['siglogMH_alf'] = np.nan

for i in range(len(SGA['SGA_ID'])):
    
    # Set the HI velocity width of this galaxy
    if SGA['ALFALFA_AGCNr'][i] >= 0:
        
        ALFALFA_idx = ALFALFA_dict[SGA['ALFALFA_AGCNr'][i]]
        
        SGA['logMH_alf'][i] = ALFALFA['logMH'][ALFALFA_idx]
        SGA['siglogMH_alf'][i] = ALFALFA['siglogMH'][ALFALFA_idx]
```

```python
# keep only the ones with HI masses
HI_mass = np.isfinite(SGA['logMH_alf'])
sga_alf_xmatch = SGA[HI_mass]
sga_alf_xmatch[:5]
```

```python
sga_alf_xmatch.write('/pscratch/sd/d/dbustos/rot_curves/sga_xmatch_HI.fits',format='fits', overwrite=True)
```
