Example code to generate Figure 11 (sky plot of Tully-Fisher galaxies in DR1) of the DR1 TF paper.

```python
from astropy.io import fits

import matplotlib.pyplot as plt
from matplotlib.projections import get_projection_names

from desiutil.plots import init_sky
```

# Data

```python
hdul = fits.open('fig11_data.fits')

alfalfa = hdul['ALFALFA'].data
desi = hdul['DESI'].data

hdul.close()
```

```python
ax = init_sky()

# ALFALFA
ax.scatter(ax.projection_ra(alfalfa['RA']), 
           ax.projection_dec(alfalfa['DEC']), 
           marker='.', 
           color='thistle', 
           label='ALFALFA')

# DESI DR1 TF "main" sample
ax.scatter(ax.projection_ra(desi['RA'][desi['MAIN']]), 
           ax.projection_dec(desi['DEC'][desi['MAIN']]), 
           marker='.', 
           color='darkblue', 
           label='DESI DR1 main')

# DESI DR1 TF removed
ax.scatter(ax.projection_ra(desi['RA'][~desi['MAIN']]), 
           ax.projection_dec(desi['DEC'][~desi['MAIN']]), 
           marker='.', 
           s=2,
           color='gray', 
           label='DESI DR1')

ax.xaxis.label.set_size(14)
ax.yaxis.label.set_size(14)

plt.legend(loc='lower right')

plt.tick_params(axis='both', which='major', labelsize=12);
```
