Example code to recreate Fig. 12 ($N(z)$ distribution) of the DESI DR1 Tully-Fisher galaxies

```python
from astropy.io import fits

import numpy as np

import matplotlib.pyplot as plt
```

# Data

```python
hdul = fits.open('fig12_data.fits')

N = hdul[1].data

hdul.close()
```

```python
# Recreate bin edges
z_bins = np.zeros(len(N['BIN_EDGE']) + 1)
z_bins[:-1] = N['BIN_EDGE']
z_bins[-1] = N['BIN_EDGE'][-1] + (N['BIN_EDGE'][1] - N['BIN_EDGE'][0])
```

# Plot

```python
plt.figure(tight_layout=True)

plt.hist(N['BIN_EDGE'], z_bins, weights=N['N_MAIN'], 
         color='darkblue', 
         label='Main Sample')
plt.hist(N['BIN_EDGE'], z_bins, weights=N['N_DWARF'], 
         color='darkgray', 
         label='Dwarfs & Outliers')
plt.hist(N['BIN_EDGE'], z_bins, weights=N['N_MAIN'], 
         color='darkblue', 
         histtype='step')

plt.legend()

plt.xlabel(r'$z_{\text{CMB}}$', fontsize=16)
plt.ylabel('number of galaxies', fontsize=16)

plt.tick_params(axis='both', which='major', labelsize=14);
```
