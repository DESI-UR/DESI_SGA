```python
from astropy.table import Table

import numpy as np
```

```python
q0 = 0.2
```

# Read in TF catalog

```python
gals = Table.read('/Users/kdouglass/Documents/Research/data/DESI/Y1/DESI-DR1_TF_pv_cat_v15.fits')

gals[:5]
```

# How many galaxies are part of the main sample that have $i < 45^\circ$?

```python
cosi2 = (gals['BA']**2 - q0**2)/(1 - q0**2)

cosi2[cosi2 < 0] = 0

i_angle = np.arccos(np.sqrt(cosi2))

np.sum(i_angle[gals['MAIN']] < (45*np.pi/180))
```
