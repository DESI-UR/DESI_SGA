This notebook generates Fig. 2 (axis ratio versus apparent magnitude) for the EDR TF paper.

```python
import numpy as np

from astropy.io import fits

import matplotlib.pyplot as plt
```

# Data

```python
hdul = fits.open('fig2_data.fits')

hdr = hdul[0].header
gals = hdul['EDR'].data
medians = hdul['MEDIANS'].data

hdul.close()
```

# Plot

```python
plt.figure(tight_layout=True)

plt.plot(gals['BA'], gals['R_MAG'], '.', alpha=0.2)
plt.errorbar(medians['BA'] + 0.05, medians['R_MAG'], yerr=medians['R_MAG_ERR'], fmt='x', c='darkblue')

plt.plot(np.array([0,1]), hdr['SLOPE']*np.array([0,1]) + hdr['YINT'], 'k')

plt.xlim([0, 1])
plt.ylim([18.5, 11.75])

plt.tick_params('both', which='major', labelsize=14)

plt.xlabel('$b/a$', fontsize=16)
plt.ylabel('$m_r$', fontsize=16);
```
