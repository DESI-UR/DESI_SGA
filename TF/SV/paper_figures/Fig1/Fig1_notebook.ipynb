Code to generate Figure 1 (sky plot of Tully-Fisher galaxies in EDR) of the EDR TF paper.

```python
from astropy.table import Table

import matplotlib.pyplot as plt
from matplotlib.projections import get_projection_names

from desiutil.plots import init_sky
```

# Data

```python
alfalfa = Table.read('fig1_alfalfa_data.txt', format='ascii.commented_header')

fuji = Table.read('fig1_EDR_data.txt', format='ascii.commented_header')
```

# Plot

```python
ax = init_sky()

ax.scatter(ax.projection_ra(alfalfa['RA']), 
           ax.projection_dec(alfalfa['DEC']), 
           marker='.', 
           color='thistle', 
           label='ALFALFA')
ax.scatter(ax.projection_ra(fuji['RA']), 
           ax.projection_dec(fuji['DEC']), 
           marker='.', 
           color='tab:blue', 
           label='DESI EDR')

ax.xaxis.label.set_size(14)
ax.yaxis.label.set_size(14)

plt.tick_params(axis='both', which='major', labelsize=12);
```
