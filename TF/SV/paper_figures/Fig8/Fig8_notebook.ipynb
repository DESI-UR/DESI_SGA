Example notebook to recreate Fig. 8 (calibrated TFR corner plot) for the EDR TF paper.

```python
import numpy as np

from astropy.io import fits

import corner
```

# Data

```python
hdul = fits.open('fig8_data.fits')

mcmc_samples = hdul[0].data

hdul.close()
```

# Plot

```python
fig = corner.corner(mcmc_samples.T, bins=30, smooth=1,
                    labels=['$a$', '$b_{Coma}$', '$b_{0pt}$', r'$\sigma_{Coma}$'],
                    label_kwargs={'fontsize':18}, 
                    labelpad=0.1, 
                    levels=(1-np.exp(-0.5), 1-np.exp(-2)),
                    quantiles=[0.16, 0.5, 0.84],
                    color='tab:blue',
                    hist_kwargs={'histtype':'stepfilled', 'alpha':0.3},
                    plot_datapoints=False,
                    fill_contours=True,
                    show_titles=True,
                    title_kwargs={"fontsize": 18, 'loc':'left', 'pad':10})

for ax in fig.get_axes():
    ax.tick_params(axis='both', which='major', labelsize=16)
```
