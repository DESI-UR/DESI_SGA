```python
from astropy.table import Table
import matplotlib.pyplot as plt
import numpy as np
```

```python
SGA_Cindex = Table.read("/global/homes/b/barnerg/SGA_Cindex.fits")
```

```python
SGA_Cindex
```

```python
gr = SGA_Cindex['g_r']
rz = SGA_Cindex['r_z']
gz = SGA_Cindex['g_z']
C59 = SGA_Cindex['C_r50_r90']
C29 = SGA_Cindex['C_r20_r90']
```

```python
print(max(gr))
print(max(rz))
print(max(gz))
print(min(gr))
print(min(rz))
print(min(gz))
```

```python
plt.scatter(gz, C59, alpha=0.1)
#plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.title('Color(g - z) v. Concentration Index(C (r50/r90))')
plt.xlabel('g - z')
plt.ylabel('C (r50/r90)')
plt.xticks(np.arange(-7, 10, 1))
plt.yticks(np.arange(0,1.1,0.1))

plt.savefig('Color(g_z)vConcentration(C_r50_r90).png')
plt.show()
plt.close()
```
