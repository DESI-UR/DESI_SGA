```python
from astropy.table import Table
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches
```

```python
SGA = Table.read('SGA_morph.fits')
```

```python
SGA
```

```python
SGA['HUBBLE_TYPE'].mask
```

```python
E = SGA[SGA['HUBBLE_TYPE']==-5.0]
```

```python
L = SGA[(SGA['HUBBLE_TYPE']==-2.0) | (SGA['HUBBLE_TYPE']==0)]
```

```python
#Sa = SGA[SGA['HUBBLE_TYPE']==1.0]
```

```python
Sb = SGA[(SGA['HUBBLE_TYPE']==2.0) | (SGA['HUBBLE_TYPE']==3.0) | (SGA['HUBBLE_TYPE']==4.0)]
```

```python
S =  SGA[(SGA['HUBBLE_TYPE']==1.0) | (SGA['HUBBLE_TYPE']==5.0) | (SGA['HUBBLE_TYPE']==6.0) | (SGA['HUBBLE_TYPE']==7.0)]
```

```python
I = SGA[SGA['HUBBLE_TYPE']==9.0]
```

```python
Nonclass = SGA[SGA['HUBBLE_TYPE'].mask]
```

```python
gr = SGA['g_r']
rz = SGA['r_z']
gz = SGA['g_z']
C59 = SGA['C_r50_r90']
C29 = SGA['C_r20_r90']
```

```python
plt.scatter(Nonclass['g_z'], Nonclass['C_r50_r90'], c='tab:gray',label='No class (grey)', alpha=0.1)
plt.scatter(E['g_z'], E['C_r50_r90'], c='#ff0000',label = 'Elliptical (red)', alpha=0.1)
plt.scatter(L['g_z'], L['C_r50_r90'], c='tab:orange', label = 'Lenticular (orange)',alpha=0.1)
plt.scatter(S['g_z'], S['C_r50_r90'], c='y', label='Unbarred Spiral (yellow)',alpha=0.1)
plt.scatter(Sb['g_z'], Sb['C_r50_r90'], c='c',label='Barred Spiral (blue)', alpha=0.1)
plt.scatter(I['g_z'], I['C_r50_r90'], c='tab:purple',label='Irregular (purple)', alpha=0.1)
legend = plt.legend(loc='upper right')


#plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.title('Color(g - z) v. Concentration Index(C (r50/r90))')
plt.xlabel('g - z')
plt.ylabel('C (r50/r90)')
plt.xticks(np.arange(-2, 4, 1))
plt.yticks(np.arange(0,1.1,0.1))

plt.savefig('Color(g_z)vConcentration(C_r50_r90).png')
plt.show()
plt.close()
```
