```python
from astropy.table import Table
import matplotlib.pyplot as plt
import numpy as np
```

```python
SGA_Cindex = Table.read('SGA_morph.fits')
```

```python
SGA_Cindex
```

```python
M0= SGA_Cindex['R_COG_PARAMS_M0']
Alpha1= SGA_Cindex['R_COG_PARAMS_ALPHA1']
Alpha2= SGA_Cindex['R_COG_PARAMS_ALPHA2']
Mtot= SGA_Cindex['R_COG_PARAMS_MTOT']
r50 = SGA_Cindex['r50']
r0= np.power(10,11)
```

```python
g = (-Alpha2)
l = (r50/r0)
a = np.power(l,g) * Alpha1
m = np.log(1 + a) * M0
mr50 = Mtot + m
```

```python
SGA_Cindex.add_column(col= mr50, name ='m(r50)')
```
