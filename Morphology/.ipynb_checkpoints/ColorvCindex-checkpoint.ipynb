```python
from astropy.table import Table
import matplotlib.pyplot as plt
import numpy as np
```

```python
Sc = SGA[(SGA['HUBBLE_TYPE']==5.0) | (SGA['HUBBLE_TYPE']==6.0) | (SGA['HUBBLE_TYPE']==7.0)]
```

```python
I = SGA[SGA['HUBBLE_TYPE']==9.0]
```

```python
gr = SGA['g_r']
rz = SGA['r_z']
gz = SGA['g_z']
C59 = SGA['C_r50_r90']
C29 = SGA['C_r20_r90']
```

```python
plt.scatter(SGA['g_r'], SGA['C_r50_r90'], c='tab:gray', alpha=0.01)
#plt.scatter(E['g_r'], E['C_r50_r90'], c='#ff0000', alpha=0.01)
#plt.scatter(L['g_r'], L['C_r50_r90'], c='m', alpha=0.01)
#plt.scatter(Sa['g_r'], Sa['C_r50_r90'], c='c', alpha=0.01)
#plt.scatter(Sb['g_r'], Sb['C_r50_r90'], c='y', alpha=0.01)
#plt.scatter(Sc['g_r'], Sc['C_r50_r90'], c='g', alpha=0.01)
#plt.scatter(I['g_r'], I['C_r50_r90'], c='b', alpha=0.01)
#plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.title('Color(g - z) v. Concentration Index(C (r50/r90))')
plt.xlabel('g - z')
plt.ylabel('C (r50/r90)')
plt.xticks(np.arange(-7, 10, 1))
plt.yticks(np.arange(0,1.1,0.1))


plt.savefig('Color(g_z)vConcentration(C_r50_r90).png')
plt.show()
plt.close()
```
